```python
import numpy as np
import matplotlib.pyplot as plt
np.random.seed(100)

def plotData(X,y):
    plt.scatter(X,y)
    plt.title('Data')
    plt.xlabel('x')
    plt.ylabel('y')
    plt.show()

n=500
x=np.random.randn(n,1)

y=2*x+1+np.random.randn(n,1)

plotData(x,y)
```

```python
ones=np.ones((n,1))
X=np.append(x,ones,axis=1)
```

```python
print(X)
```

```python
w=np.dot(np.linalg.inv(np.dot(X.T,X)),np.dot(X.T,y))
```

```python
print(w)
```

```python
y_hat=np.dot(X,w)
```

```python
print(y_hat)
```

```python
def plotData1(x,y,y_hat=None):
    plt.scatter(x,y,marker='.')
    if y_hat is not None:
      plt.scatter(x,y_hat,marker='s')
    plt.show()
```

```python
plotData1(x,y,y_hat)
```
