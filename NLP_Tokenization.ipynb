```python
import pandas as pd
import numpy as np

from google.colab import drive
drive.mount('/content/drive')

dataset_file = '/content/drive/MyDrive/ML/Sinhala_Singlish_Hate_Speech.csv'
data=pd.read_csv(dataset_file)
```

```python
import nltk as nk
nk.download('punkt')
combined_text = data.apply(lambda row: ' '.join(row.astype(str)), axis=1).str.cat(sep='.')
tokens=nk.word_tokenize(combined_text)
print(tokens)
```

```python
nk.download('averaged_perceptron_tagger')
tagged=nk.pos_tag(tokens)
print(tagged)
```
